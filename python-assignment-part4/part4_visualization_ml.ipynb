```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load Dataset 
df = pd.read_csv("students.csv")
df.head()
```

```python
#Bar Plot
plt.figure(figsize=(6,4))
plt.bar(df['name'], df['math'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("plot1_bar.png")
plt.close()
```

```python
#Histogram
plt.figure(figsize=(6,4))
plt.hist(df['math'], bins=5)
plt.tight_layout()
plt.savefig("plot2_hist.png")
plt.close()
```

```python
#Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(df['math'], df['science'])
plt.xlabel("Math")
plt.ylabel("Science")
plt.tight_layout()
plt.savefig("plot3_scatter.png")
plt.close()
```

```python
#Box Plot
plt.figure(figsize=(6,4))
plt.boxplot(df['math'])
plt.tight_layout()
plt.savefig("plot4_box.png")
plt.close()
```

```python
#Line Plot
plt.figure(figsize=(6,4))
plt.plot(df['name'], df['math'], marker='o')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("plot5_line.png")
plt.close()
```

```python
#Seaborn Bar Plot
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.barplot(data=df, x='passed', y='math')

plt.subplot(1,2,2)
sns.barplot(data=df, x='passed', y='science')

plt.tight_layout()
plt.savefig("plot6_seaborn_bar.png")
plt.close()
```

```python
#Seaborn Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='math', y='science', hue='passed')
plt.tight_layout()
plt.savefig("plot7_seaborn_scatter.png")
plt.close()
```
